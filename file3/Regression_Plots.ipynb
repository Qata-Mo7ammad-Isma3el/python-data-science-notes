**Regression_Plots**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels 

sns.set_style('darkgrid')
tips=sns.load_dataset('tips')
tips
```

```python
sns.regplot(data=tips , x='total_bill' , y='tip')
plt.tight_layout()
plt.show()
```

```python
sns.regplot(data=tips , x='size' , y='tip' , x_jitter=0.1 , x_estimator=np.mean)
#you can block a regression line between categorical variable and numerical one
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20 , 20))
sns.lmplot(data=tips , x='total_bill' , y='tip' , hue='smoker', palette='autumn' , 
            col='smoker' , row='sex' , truncate=False)
plt.show()
```

```python
tips['smoker'] = tips['smoker'] == 'Yes'
#sns.lmplot(data=tips , x='total_bill' , y='smoker' , logistic=True)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue="smoker", col="smoker", 
            scatter_kws={'s': 100, 'edgecolor':'black'}, 
            line_kws={'lw': 6, 'ls': '--'})
plt.show()
```
