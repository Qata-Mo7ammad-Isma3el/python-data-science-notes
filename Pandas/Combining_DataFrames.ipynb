**there are 4 types of join**
1) `Inner join`: Inner join in pandas is used to merge two data frames at the intersection. It returns the mutual rows of both.

2) `Outer join`: It might also be called Full Outer Join and returns all rows from both DataFrames. It will match all rows that exist in both data frames. The rows found only in one of the two DataFrames will get the `NAN` value.

3) `Left join`: The left outer join returns all rows from the left data frame, which will be merged with the corresponding rows from the right data frame. It fills the unmatched rows with `NAN`.

4) `Right join`: The right outer join returns all rows from the right data frame and the remaining data from the left. The data which does not correspond to the right data frame will have `NANs` assigned.

`Inner join`

```python
x=pd.DataFrame({'cat_name':['Elec' ,'Games', 'Kitchen'] , 'cat_id':[0 , 1 , 2]})
y=pd.DataFrame({'item_name':['Macbook' ,'Panting' , 'Cup'] , 'item_id':[0 , 1 , 2] , 'cat_id':[0 , 5 ,2]})
print('x data frame:\n',x , end='\n\n')
print('y data frame:\n',y , end='\n\n')
merged_data=pd.merge(x , y , how='inner' , on='cat_id')
print('join of type inner on cat_id:\n',merged_data, end='\n\n')
```

`Outer join`

```python
merged_data=pd.merge(x , y , how='outer' , on='cat_id')
print('join of type outer on cat_id:\n',merged_data, end='\n\n')
```

`Left join`

```python
merged_data=pd.merge(x , y , how='left' , on='cat_id')
print('join of type outer on cat_id:\n',merged_data, end='\n\n')
```

`Right join`

```python
merged_data=pd.merge(x , y , how='right' , on='cat_id')
print('join of type outer on cat_id:\n',merged_data, end='\n\n')
```

**some important parameters**

```python
right=pd.DataFrame({'key_right':['foo' , 'bar'] , 'val':[1 , 2]})
left=pd.DataFrame({'key_left':['foo' , 'bar'] , 'val':[3 , 4]})
pd.merge(left , right , how='inner' ,right_on='key_right' , left_on='key_left', suffixes=('_from_left' , '_from_right'))
#alternative way
left.merge(right , how='inner' ,right_on='key_right' , left_on='key_left', suffixes=('_from_left' , '_from_right'))
```

`join()`: method as same as `merge()` method but `join()` will combine the data frames on the indices

```python
right=pd.DataFrame({'key':['foo' , 'bar'] , 'val':[1 , 2]})
left=pd.DataFrame({'key':['foo' , 'bar'] , 'val':[3 , 4]})

left.join(right , how='outer' , lsuffix='_from_left' , rsuffix='_from_right')
```

```python
# as you can see the join() method will combine them on the indices to change this we will use 
left=left.set_index('key' )
right=right.set_index('key')
left.join(right , how='outer' , lsuffix='_from_left' , rsuffix='_from_right')
```

`concat()` and **some important parameters**

```python
df_3=pd.DataFrame({
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
} , index=[0,1,2,3])

df_4=pd.DataFrame({
    'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
} , index=[4 , 5 , 6 ,7])
pd.concat([df_3 , df_4], keys=['first' , 'second'])
#pd.concat([df_3 , df_4], keys=['first' , 'second']).reset_index()
```
