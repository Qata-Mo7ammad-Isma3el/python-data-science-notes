**Distribution Plots**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(penguins, x='flipper_length_mm' , bins=20 , stat='density')
plt.show()
```

```python
sns.histplot(penguins, x='species' , bins=20 , stat='density' , shrink=0.8)
plt.show()
```

```python
sns.histplot(penguins, x='flipper_length_mm' , bins=20 , stat='density',hue='species' ,
                element='step' , multiple='stack' , kde=True  , common_norm=False)
plt.show()
```

```python
sns.histplot(penguins, x='flipper_length_mm',y='bill_length_mm' , bins=20 , stat='density',hue='species' ,
                element='step' , multiple='stack' , kde=True  , common_norm=False)
plt.show()
```

```python
sns.histplot(penguins, x='flipper_length_mm' , bins=20 , stat='density',hue='species' ,
                element='step' , multiple='stack' , kde=True  , common_norm=False)
plt.show()
```
