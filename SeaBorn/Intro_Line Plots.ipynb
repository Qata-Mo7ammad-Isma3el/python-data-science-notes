`Seaborn` is a developed library based on `matplotlib` have some more features and the same use it visualize data 

**Types of plots in `Seaborn`**

1) relational plots: `sns.relplot`
2) categorical plots: 
3) Distribution plots

**SET UP**

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
import PyQt5

flights=sns.load_dataset('flights')
flights
```

```python
Feb_flights=flights[flights['month']=='Feb']

sns.set_style('darkgrid')
sns.lineplot(x='year' , y='passengers' , data=Feb_flights)

# as you can see x-axis label and y-axis label are defined by default 
# but you can still use matplotlib to edit your graph

plt.xlabel('Year')
plt.title('Year vs. Passengers')
plt.show()
```

```python
%matplotlib inline 
sns.set_style('darkgrid')
plt.figure(figsize=(5 , 5))
sns.lineplot(x='year' , y='passengers' , data=flights ,style='month', size='month', hue='month')

# to plot all data using seaborn it will calculate the mean of all months in the year and plot it 
# and the shaded area is simply the 95th confidence interval
plt.xlabel('Year')
plt.title('Year vs. Passengers')
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.lineplot(data=fmri , x='timepoint' ,y='signal' , hue='event' , style='region')
plt.show()
```

```python
sns.relplot(data=fmri , x='timepoint' ,y='signal' , hue='event' , 
            style='region' , kind='line' , col='event' , row='region')
plt.tight_layout()
plt.show()
```
