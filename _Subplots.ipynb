**Subplots**

**important description**
`plt.subplots()`: used to plot more than one plot in one figure it will return two objects:

1) `fig`: stands for all the figure
2) `axes`: it will be the new object and each index represent a plot space 

**important parameters**

`nrows = 2`: number of row 

`ncols = 2`:number of column

`sharex = True`  and `sharey=True` : if your data have the same x or y values you can share the axis on the plots

`figsize = (10,10)`:to control and manipulate size of plots

$note  that:$

`plt.title()`  **=>** `axis[row,colum].set_title()`

`plt.xlabel()` **=>** `axis[row,colum].set_xlabel()`

`plt.ylabel()` **=>** `axis[row,colum].set_ylabel()`

`plt.savefig('image.png')` **=>** `fig.savefig('salaryage.png')`
