**Histograms**

`value_count()`: very very important method to count the data  

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline 
#plt.style.use('fivethirtyeight')
ages = [10, 30, 45, 30, 31, 46, 12, 16, 32, 23, 28, 29]
bins = [10,20,30,40,50,60]

plt.hist(ages ,bins=bins, edgecolor='black' , linewidth=2)
plt.axvline(np.mean(ages) , color='r' , label='mean')


plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.legend()
plt.show()
```

```python
data = pd.read_csv('500_Person_Gender_Height_Weight_Index.csv')
bins=[i for i in range(140 , 201 , 5)]

plt.hist(data['Height'] ,bins=bins , edgecolor='black' , linewidth=2)


plt.title('Hight Distribution')
plt.xlabel('Hight')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**exercise**

```python
Gender_Frequency=data['Gender'].value_counts()
Gender_Frequency.index

plt.bar(Gender_Frequency.index ,Gender_Frequency ,color=['r' , 'g'],
        edgecolor='black' , linewidth=2)
plt.title('Gender Bar chart')
plt.xlabel('Gender')
plt.ylabel('Frequency')

plt.show()
```
