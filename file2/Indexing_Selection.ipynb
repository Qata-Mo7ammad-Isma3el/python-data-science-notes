### **Index based selection**
`iloc` and `loc`

Select the first row of the data

```python
#iloc method works with integer indexes regardless if the data have names or any thing else as indexes 
df.iloc[0,:]
df.loc[0 , :]
```

Select the first column

```python
df.iloc[: , 0]
df.loc[: , 'Review #']
```

select the first 3 rows

```python
df.iloc[0:3, :] # same as df.head(3)
df.loc[0:2, :] # note that the end in loc is in the interval
```

select the last column

```python
df.iloc[: , -1]
```

Fancy indexing

```python
df.iloc[[45 , 34 , 10] ,: ]
df.loc[[45 , 34 , 10] ,: ]
```

select the last 10 rows

```python
df.iloc[-10:,:]
```

Select `Brand` , `Style` and `Stars` columns

```python
df.loc[: , ['Brand' , 'Style' , 'Stars']] # Fancy indexing
```

the difference between `iloc` and `loc` 

1) `iloc` : works in the implicit indexing 
2) `loc` :  works in the explicit indexing

```python
df.iloc[0:4 , :]
# indexing is [a , b)
```

```python
df.loc[0:4 , :] # note that the end in loc is in the interval
# indexing is [a , b]
```

`set_index()`

```python
df_1=df.set_index('Review #')
# the idea here that we have take the column from the same data frame and put it as indexes 
df_1
```

```python
df_1.iloc[0:3 , :]
```

```python
df_1.loc[2580:2578 , :]
```
