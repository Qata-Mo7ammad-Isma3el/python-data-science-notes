```python
import pandas as pd
import numpy as np
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic['sex'].value_counts()
titanic['who'].value_counts()
```

`pivot_table()`

```python
titanic.pivot_table(index='who' , columns='class' , values='age' , aggfunc='mean' ,observed=False )
```

```python
titanic.pivot_table(index='sex' ,columns='class' ,values='survived' , aggfunc='count' , observed=False)
```

```python
titanic.pivot_table(index='sex' ,columns='class' ,values=['survived' , 'age'] , 
                    aggfunc={'age':'mean','survived':'count'} , observed=False)
```
