```python
import pandas as pd
from matplotlib import pyplot as plt 
import numpy as np
print(plt.style.available)
plt.style.use('seaborn-v0_8')
data=pd.read_csv('C:/Users/Mohammad Ismail/Desktop/ML with python/intro_MatPlotLib/house prices/train.csv')
```

**Lets play with some simple data first before this data frame** 

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 10, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
            538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]


plt.scatter(x , y  , c=colors,marker='.' ,edgecolors='black' , sizes=sizes, linewidths=1 , alpha=0.7 , cmap='Reds')

plt.xlabel('x data')
plt.ylabel('y data')
plt.title('first scatter plot')
cbar=plt.colorbar()
cbar.set_label('Satisfaction')
plt.show()
```

2) methods:

`cber=plt.colorbar()`: this method will give you a color bar next to the `Scatter` simply it will give more information's about your data it will become clear in parameter section 

`cbar.set_label('some label')`: to set a label to the color bar 

3) parameters:

$note that:$ if you have information's you want to show it using the point colors you can't use `color=` parameter it will arise an error instead you have to use `c=` parameter


`c=colors`: if you want to show more information's about your data this option will preform it perfectly `colors` is a `list` than contain data. `scatter()` will use it as Color intensity  


`marker='.'`: we have seen this in previous codes


`edgecolors='black'`: give you a line to clearly separate points on edges 

`linewidths=1`: simply it will sets the line width of the edges on each point  


`sizes=sizes`:if you want to show more information's about your data this option will preform it perfectly `sizes` is a `list` than contain data. `scatter()` will use it as point sizes  

`alpha=0.75`:Color intensity

`cmap='Reds'`: to set styles 

`plt.xscale('log')` and `plt.yscale('log')` : to normalize the data and deal with outliers




**Lets now get back to the data frame** 

```python

plt.scatter(data['YearBuilt'] , data['SalePrice'] , c=data['OverallQual'] , sizes=15*data['OverallCond']
            , alpha=0.75, cmap='tab10')
cbar=plt.colorbar()
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
cbar.set_label('Quality')
plt.tight_layout()
plt.show()
```
