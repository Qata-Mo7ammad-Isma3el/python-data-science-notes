**Pi Charts**

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

# Language Popularity
# slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 
#             18017, 7920, 7331, 7201, 5833]
# labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++',
#             'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0 , 0 , 0 , 0.1 , 0]
plt.pie(slices , labels=labels ,explode=explode, shadow=True,startangle=90, autopct='%1.1f%%', 
        rotatelabels=True, wedgeprops={'edgecolor': 'black'})

plt.title('My Pie Chart')
plt.tight_layout()
plt.show()

```

**some important parameters**

`plt.pi(slices , labels=labels)`: to plot a pi chart simply pass `slices` and `labels`

`explode=explode`: `explode` is a python `list` that have the same length of labels list and it has values between $[0, 1]$ for each label 0 means the slice will remain the same and $(0 , 1]$ will explode the slice out 

`shadow=True`: give slices a beautiful shadow the default value is `False`

`startangle=80`: preform a rotation on your pi  

`autopct='%1.1f%%'`: calculate the percentage of each slice

`rotatelabels=True`:preform a rotation on your labels

`wedgeprops={'edgecolor': 'black'}`: give you a line to clearly separate slices on edges
